```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
from matplotlib.dates import MonthLocator, DateFormatter
from google.colab import drive
drive.mount('/content/drive/')

# read the two CSV files into pandas dataframes
df1 = pd.read_csv('/content/drive/My Drive/Sales_data.csv')
df2 = pd.read_csv('/content/drive/My Drive/Sales_data_2122.csv')
```

```python
# Merge the two dataframes using the concat() function
merged_sales_df = pd.concat([df1, df2])

# Write the merged dataframe to a new CSV file
merged_sales_df.to_csv('/content/drive/My Drive/merged_sales_data.csv', index=False)
sales_data = pd.read_csv('/content/drive/My Drive/merged_sales_data.csv')
```

```python
sales_data['sold_qty'] = sales_data['sold_qty'].abs()
```

```python
# filter out rows corresponding to February 2023
sales_data['bill_date'] = pd.to_datetime(sales_data['bill_date'])
sales_data = sales_data[(sales_data['bill_date'].dt.month != 2) | (sales_data['bill_date'].dt.year != 2023)]
```

```python
monthly_sales = sales_data.groupby([sales_data['bill_date'].dt.year.rename('year'), sales_data['bill_date'].dt.month.rename('month')])['sales_rate'].sum()
monthly_sales.head()
```

```python
import matplotlib.pyplot as plt

monthly_sales.plot(kind='line', figsize=(15, 5))
plt.title('Monthly Sales')
plt.xlabel('Year, Month')
plt.ylabel('Sales (INR)')
plt.grid(True)
plt.show()
```

```python
sales_data['bill_date'] = pd.to_datetime(sales_data['bill_date'])


# Filter data for last 18 months
last_18_months = sales_data[sales_data['bill_date'] >= (sales_data['bill_date'].max() - pd.DateOffset(months=23))]

# Group daily sales data into monthly sales data
monthly_sales_data = last_18_months.groupby(pd.Grouper(key='bill_date', freq='MS')).agg({'sold_qty': 'sum', 'sales_rate': 'mean'}).reset_index()

# Calculate total revenue (sales) for each month
monthly_sales_data['total_revenue'] = monthly_sales_data['sold_qty'] * monthly_sales_data['sales_rate']

# Prepare X (features) and y (target) for training
X = monthly_sales_data[['sold_qty', 'sales_rate']]
y = monthly_sales_data['total_revenue']

# Train the Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Predict future sales for the next month


next_month =  pd.Timestamp(2023, 2, 1)
future_data = monthly_sales_data.append({'bill_date': next_month}, ignore_index=True)








# Fill data for future_data using the data available in the training dataframe
future_data['sold_qty'] = future_data['sold_qty'].fillna(monthly_sales_data['sold_qty'].mean())
future_data['sales_rate'] = future_data['sales_rate'].fillna(monthly_sales_data['sales_rate'].mean())



# Make predictions for the next month
future_data['total_revenue'] = model.predict(future_data[['sold_qty', 'sales_rate']])


# Fill data for future_data using the data available in the training dataframe
future_data['sold_qty'] = future_data['sold_qty'].fillna(method='ffill')  # Forward fill missing values
future_data['sales_rate'] = future_data['sales_rate'].fillna(method='ffill')  # Forward fill missing values

# Make predictions for the next 3 months
future_data['total_revenue'] = model.predict(future_data[['sold_qty', 'sales_rate']])
print(future_data[['bill_date', 'total_revenue']])

# Set float format to display values with two decimal places
pd.options.display.float_format = '{:.2f}'.format



# Calculate accuracy of the model
y_true = monthly_sales_data['total_revenue']
y_pred = model.predict(X)
r2 = r2_score(y_true, y_pred)
print("R2 Score:", r2)

# Plot actual vs predicted sales
plt.figure(figsize=(15, 5))
plt.plot(monthly_sales_data['bill_date'], monthly_sales_data['total_revenue'], label='Actual Sales')
plt.plot(monthly_sales_data['bill_date'], model.predict(X), label='Predicted Sales')
plt.plot(future_data['bill_date'], future_data['total_revenue'], 'r--', label='Next Month Prediction')

# # Add month labels to x-axis
# months = MonthLocator()
# months_fmt = DateFormatter('%b %Y')
# plt.gca().xaxis.set_major_locator(months)
# plt.gca().xaxis.set_major_formatter(months_fmt)

plt.xlabel('Bill Date')
plt.ylabel('Total Revenue')
plt.title('Actual vs Predicted Sales')
plt.legend()
plt.show()
```

```python
# Prepare X (features) and y (target) for training
X = monthly_sales_data[['sold_qty', 'sales_rate']]
y = monthly_sales_data['total_revenue']

# Train the Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Predict future sales for the next 3 months
next_months = pd.date_range(monthly_sales_data['bill_date'].max() + pd.DateOffset(months=1), periods=3, freq='MS')

future_data = pd.DataFrame({'bill_date': next_months})
for i, row in future_data.iterrows():
    # Get data from same month last year
    last_year_month = row['bill_date'] - pd.DateOffset(years=1)
    last_year_data = monthly_sales_data[monthly_sales_data['bill_date'] == last_year_month]
    if not last_year_data.empty:
        sold_qty = last_year_data['sold_qty'].values[0]
        sales_rate = last_year_data['sales_rate'].values[0]
        future_data.loc[i, 'sold_qty'] = sold_qty
        future_data.loc[i, 'sales_rate'] = sales_rate

# Fill any remaining missing values with the mean
future_data = future_data.fillna(monthly_sales_data.mean())

# Make predictions for the next 3 months
future_data['total_revenue'] = model.predict(future_data[['sold_qty', 'sales_rate']])

# Print the predicted revenue for the next 3 months
print(future_data[['bill_date', 'total_revenue']])

# Plot actual vs predicted sales
# Add month labels to x-axis

plt.figure(figsize=(25, 5))
plt.plot(monthly_sales_data['bill_date'], monthly_sales_data['total_revenue'], label='Actual Sales')
plt.plot(monthly_sales_data['bill_date'], model.predict(X), label='Predicted Sales')
plt.plot(future_data['bill_date'], future_data['total_revenue'], 'r--', label='Next 3 Months Prediction')

months = MonthLocator()
months_fmt = DateFormatter('%b %Y')
plt.gca().xaxis.set_major_locator(months)
plt.gca().xaxis.set_major_formatter(months_fmt)
plt.xlabel('Bill Date')
plt.ylabel('Total Revenue')
plt.title('Actual vs Predicted Sales')
plt.legend()
plt.show()
```
